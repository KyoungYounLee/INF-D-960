```python
import graphviz as gv

dot = gv.Digraph(comment='Relational Algebra')
dot.attr(rankdir='TB')

dot.node_attr.update(style='filled', fillcolor='white', shape='plain')

dot.node('R', '<<FONT POINT-SIZE="20">π </FONT> s.name, e.course>')
dot.node('T', '<<FONT POINT-SIZE="20">⨝ </FONT> e.grade≥avg+1>')
dot.node('left', '<<FONT POINT-SIZE="20">⨝ </FONT> s.id=e.sid>')
dot.node('left_1', '<<FONT POINT-SIZE="20">σ </FONT> s.major="CS" ∨ s.major="Games Eng">')
dot.node('students', 'students s')
dot.node('exams', 'exams e')
dot.node('exams2', 'exams e2')
dot.node('right', '<<FONT POINT-SIZE="20">Γ </FONT> ∅;avg: avg(e2.grade)>')
dot.node('right2', '<<FONT POINT-SIZE="20">σ </FONT> s.id=e2.sid ∨ (e2.curriculum=s.major ∧ s.year&gt;e2.date)>')

dot.edge('R', 'T', arrowhead='none')
dot.edge('T', 'left', arrowhead='none')
dot.edge('T', 'right', arrowhead='none')

dot.edge('left', 'left_1', arrowhead='none')
dot.edge('left_1', 'students', arrowhead='none')
dot.edge('left', 'exams', arrowhead='none')

dot.edge('right', 'right2', arrowhead='none')
dot.edge('right2', 'exams2', arrowhead='none')

dot
```

```python
dot4 = gv.Digraph(comment='Relational Algebra')
dot4.attr(rankdir='TB')

dot4.node_attr.update(style='filled', fillcolor='white', shape='plain')

dot4.node('R', '<<FONT POINT-SIZE="20">π </FONT> s.name, e.course>')
dot4.node('T',
          '<<FONT POINT-SIZE="20">⨝ </FONT> e.grade≥avg+1 ∧ (s.id=d.id ∧ s.major=d.major ∧ s.year=d.year)>')
dot4.node('left', '<<FONT POINT-SIZE="20">⨝ </FONT> s.id=e.sid>')
dot4.node('left_1', '<<FONT POINT-SIZE="20">σ </FONT> s.major="CS" ∨ s.major="Games Eng">')
dot4.node('students', 'students s')
dot4.node('exams', 'exams e')
dot4.node('exams2', 'exams e2')

dot4.node('right', '<<FONT POINT-SIZE="20">⨝</FONT>>')
dot4.node('right_left', '<<FONT POINT-SIZE="20">π </FONT> d.id:s.id, d.major:s.major, d.year:s.year>')

dot4.node('right_right', '<<FONT POINT-SIZE="20">Γ </FONT> d.id, d.major, d.year; ∅;avg: avg(e2.grade)>')
dot4.node('right_right2', '<<FONT POINT-SIZE="20">σ </FONT> d.id=e2.sid ∨ (e2.curriculum=d.major ∧ d.year&gt;e2.date)>')

with dot4.subgraph() as s:
    s.attr(rank='same')
    s.node('right_left')
    s.node('exams2')

dot4.edge('R', 'T', arrowhead='none')
dot4.edge('T', 'left', arrowhead='none')
dot4.edge('T', 'right_right', arrowhead='none')
dot4.edge('right', 'right_left', arrowhead='none')
dot4.edge('right', 'exams2', arrowhead='none', constraint='false')

dot4.edge('right_right', 'right_right2', arrowhead='none')
dot4.edge('right_right2', 'right', arrowhead='none')

dot4.edge('right_left', 'left', arrowhead='none')

dot4.edge('left', 'left_1', arrowhead='none')
dot4.edge('left_1', 'students', arrowhead='none')
dot4.edge('left', 'exams', arrowhead='none')

dot4.edge('right_left', 'exams2', style='invis')

dot4
```
